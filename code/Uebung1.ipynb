## Aufgabe1.

$(Au)(t) = \int_{0}^{1} e^{t-s} u(s) ds, \qquad t \in [0,1] $


```python
s = np.linspace(0,1,100)
u_s = s**2
plt.figure(figsize=(8, 6))
plt.xlabel('s')
plt.ylabel('u(s)')
plt.plot(s,u_s, label = r'$u(s) = s^2$')
plt.legend()
```

```python
s = np.linspace(0,1,100)
g_s = np.exp(0-s)*s**2
plt.figure(figsize=(8, 6))
plt.xlabel('s')
plt.ylabel('g(s)')
plt.plot(s,g_s, label = r'$e^{-s}s^2$')
plt.legend()
```

```python
s = np.linspace(0,1,100)
t = np.linspace(0,1,100)
S,T = np.meshgrid(s,t)
g_s = np.exp(T-S)*S**2

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.view_init(15, -110)
ax.plot_surface(S, T, g_s, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_xlabel('s')
ax.set_ylabel('t')
ax.set_zlabel('g(s,t)')
```

```python
t = np.linspace(0,1,100)
u_t = (2*np.exp(1)-5)*np.exp(t-1)
plt.figure(figsize=(8, 6))
plt.xlabel('t')
plt.ylabel('u(t)')
plt.plot(t,u_t, label = r'$(2e-5)e^{t-1}$')
plt.legend()
```
