# Visualization of strong convexity

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1
alpha = 1
beta = 5
y = np.linspace(-2,2,100)
fy = (a*y)**2
gy = 2*a*y
```

### Conditions:
 - Convexity: $f(y) \ge f(x) + \nabla f(x)^T(y - x)$
 - $\beta$-Smoothness: $f(y) \le f(x) + \nabla f(x)^T(y - x) + \frac{\beta}{2} ||x-y||^2$
 - $\alpha$-Strong Convexity: $f(y) \ge f(x) + \nabla f(x)^T(y - x) + \frac{\alpha}{2} ||x-y||^2$
 
The following plot shows the listed conditions evaluated at a reference infex with respect to a quadratic function $f(y)$. 

```python
ref_idx = 35
plt.figure(figsize=(8, 6))
plt.plot(y,fy,'-',label='$f(y) = (ay)^{2}$')
plt.plot(y[ref_idx],fy[ref_idx],'.', markersize=12, label=r'$x_{ref} = $' + "%.2f" % y[ref_idx])
plt.plot(y,fy[ref_idx] + gy[ref_idx]*(y-y[ref_idx]),'-',label= r"$g_1(y) =f(x_{ref}) + \nabla f(x_{ref})(y - x_{ref})$ (Convexity)")
plt.plot(y,fy[ref_idx] + gy[ref_idx]*(y-y[ref_idx]) + beta/2*(y[ref_idx] - y)**2,'-',label= r"$g_2(y) = f(x_{ref}) + \nabla f(x_{ref})\
         (y - x_{ref}) + \frac{\beta}{2} ||x_{ref}-y||^2$ ($\beta$-Smoothness)")
plt.plot(y,fy[ref_idx] + gy[ref_idx]*(y-y[ref_idx]) + alpha/2*(y[ref_idx] - y)**2,'-',label= r"$g_3(y) = f(x_{ref}) + \nabla f(x_{ref})\
         (y - x_{ref}) + \frac{\alpha}{2} ||x_{ref}-y||^2$ ($\alpha$-Convexity)")
plt.xlabel('y')
plt.legend()
```
