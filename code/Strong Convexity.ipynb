# Visualization of strong convexity

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1
alpha = 1
beta = 5
x = np.linspace(-2,2,100)
y = np.linspace(-2,2,100)
fx = (a*x)**2
fy = (a*y)**2
gx = 2*a*x
```

### Visualization of 
 - Convexity: $f(y) \ge f(x) + \nabla f(x)^T(y - x)$
 - $\beta$-Smoothness: $f(y) \le f(x) + \nabla f(x)^T(y - x) + \frac{\beta}{2} ||x-y||^2$
 - $\alpha$-Strong Convexity: $f(y) \ge f(x) + \nabla f(x)^T(y - x) + \frac{\alpha}{2} ||x-y||^2$
 

```python
x_ref = 25
plt.figure(figsize=(8, 6))
plt.plot(y,fy,'-',label='$f(y) = (ax)^{2}$')
plt.plot(y[x_ref],fy[x_ref],'.', markersize=12, label=r'$x_{ref}$')
plt.plot(x,fx[x_ref] + gx[x_ref]*(y-x[x_ref]),'-',label= r"$f(x_{ref}) + \nabla f(x_{ref})(y - x_{ref})$ (Convexity)")
plt.plot(x,fx[x_ref] + gx[x_ref]*(y-x[x_ref]) + beta/2*(x[x_ref] - y)**2,'-',label= r"$f(x_{ref}) + \nabla f(x_{ref})\
         (y - x_{ref}) + \frac{\beta}{2} ||x_{ref}-y||^2$ ($\beta$-Smoothness)")
plt.plot(x,fx[x_ref] + gx[x_ref]*(y-x[x_ref]) + alpha/2*(x[x_ref] - y)**2,'-',label= r"$f(x_{ref}) + \nabla f(x_{ref})\
         (y - x_{ref}) + \frac{\alpha}{2} ||x_{ref}-y||^2$ ($\alpha$-Convexity)")
plt.xlabel('y')
plt.legend()
plt.savefig('Strong_convexity.png')
```
